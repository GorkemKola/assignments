Question 5) Please read the "Hotels1.csv" and "Hotels2.csv" files as a DataFrame object. Print the length and the first 5 tuples of them.

```python
#codes here
hotels_df = pd.concat([hotels1,hotels2],axis=1)
hotels_df = hotels_df.drop(["name"], axis=1)
hotels_df.head(10)
```

Question 8) Write records stored in a DataFrame that is given to you as image to a SQLite database and read all data from SQL via a SQL query


```python
#codes here
conn = connect(':memory:') #Create the database in RAM
df = {
    "course_codes" : ["AIN101", "AIN212","AIN212","AIN311","AIN440","BBM201","BBM301","BBM301"],
    "course_name" : ["Freshman Seminar", "Elements of Data Science", "Elements of Data Science", "Foundations of Machine Learning", "Introduction to Natural Language Processing", "Data Structers", "Programming Languages","Programming Languages"],
    "day" : ["Monday","Tuesday","Monday","Wednesday","Tuesday","Friday","Thursday","Thursday"],
    "number_of_students" : [80,35,35,40,25,43,70,70]
}
DATAFRAME = pd.DataFrame(df)
DATAFRAME.to_sql("DATAFRAME",conn)
pd.read_sql("SELECT * FROM DATAFRAME",conn)
```

Question 9) Filter courses that are hold on Tuesday and number of students are higher than 30 by using a SQL query

```python
#codes here
pd.read_sql("SELECT * FROM DATAFRAME WHERE day == 'Tuesday' AND number_of_students > 30 ",conn)
```

Question 10) Show the total number of students by using a SQL query

```python
#codes here
pd.read_sql("SELECT SUM(number_of_students) FROM DATAFRAME",conn)
```

Question 11) Show course names in ascending order by using a SQL query

```python
#codes here
pd.read_sql("SELECT course_name FROM DATAFRAME ORDER BY course_name",conn)
```

Question 12) Show courses which codes 'AIN 212' by using a SQL query

```python
#codes here
pd.read_sql("SELECT * FROM DATAFRAME WHERE course_codes = 'AIN212'",conn)
```

Question 13)  Show all data without duplicates by using a SQL query

```python
#codes here
pd.read_sql("SELECT DISTINCT course_codes,course_name,day,number_of_students FROM DATAFRAME",conn)
```

Question 14) Change AIN311 course's day to Friday and display it by using a SQL query
